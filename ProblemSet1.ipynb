element - 0 : 25
element - 1 : 47
element - 2 : 42

Expected Output:
The Even elements are:
42 56 32
The Odd elements are :
25 47

```python
using System.IO;
using System;

class Program
{
    static void Main()
    {
        Console.WriteLine("Enter the five elements");
        int[] numbers = new int[5];
        int[] evenNumbers = new int[5];
        int[] oddNumbers = new int[5];
        for (int i = 0; i < 5; i++)
        {
            Console.WriteLine("element - " + (i+1));
            numbers[i] = Convert.ToInt32(Console.ReadLine());
        }
        for (int i = 0; i < 5; i++)
        {
            if (numbers[i] % 2 == 0)
            {
                for (int j = 0; j < evenNumbers.Length; j++)
                {
                    if (evenNumbers[j] == 0)
                    {
                        evenNumbers[j] = numbers[i];
                        break;
                    }
                }
            }
            else
            {
                for (int j = 0; j < oddNumbers.Length; j++)
                {
                    if (oddNumbers[j] == 0)
                    {
                        oddNumbers[j] = numbers[i];
                        break;
                    }
                }
            }
        }
        Console.WriteLine("Odd Numbers are");
        foreach (int x in oddNumbers)
        {
            if (x == 0)
                break;
                Console.Write(x+"\t");
        }
        Console.WriteLine("\nEven Numbers are");
        foreach (int x in evenNumbers)
        {
            if (x == 0)
                break;
            Console.Write(x+"\t");
        }

    }
}
```

# Question 6
You can turn a word into pig-Latin using the following two rules (simplified):
• If the word starts with a consonant, move that letter to the end and append
'ay'. For example, 'happy' becomes 'appyhay' and 'pencil' becomes 'encilpay'.
• If the word starts with a vowel, simply append 'way' to the end of the word.
For example, 'enter' becomes 'enterway' and 'other' becomes 'otherway' . For
our purposes, there are 5 vowels: a, e, i, o, u (so we count y as a consonant).
Write a function pig() that takes a word (i.e., a string) as input and returns its pigLatin form. Your function should still work if the input word contains upper case
characters. Your output should always be lower case however.

```python
def pig(S):
    S=S.lower()
    vowelsList = ['a','e','i','o','u']
    if S[0] in vowelsList:
        S=S+"way"
        print(S)
    else:
        newStr=S[1:]+S[0]+"ay"
        print(newStr)
temp=True
while(temp):
    name = input("Enter the word that needs pigLatin conversion and press type exit to stop execution\t")
    if (name.lower()=='exit'):
        temp = False;
    else:
        pig(name)
```

# Question 7
File bloodtype1.txt records blood-types of patients (A, B, AB, O or OO) at a clinic.
Write a function bldcount() that reads the file with name name and reports (i.e.,
prints) how many patients there are in each bloodtype. 

```python
def bldcount(filename):
    file1 = open(filename,"r")
    bloodTypeList = ['A','AB','B','O','OO']
    words = file1.read().split()
    for i in bloodTypeList:
        count=0
        for j in words:
            if i == j:
                count=count+1
        if count ==0:
            print("There are no patients for blood type",i)
        elif count ==1:
            print("There is one patient for blood type", i)
        elif count>1:
            print("There are", count, "patients for blood type", i)
name = input("enter the file name that needs to be processed\t ")
bldcount(name)
```
